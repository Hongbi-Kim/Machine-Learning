# 문제 1-1
- crim 항목의 상위에서 10번째 값(즉, 상위 10번째 값 중에서 가장 작은 값)으로 상위 10개의 값을 변환하고 age 80 이상인 값에 대하여 crim 평균을 구하라.
- 데이터셋 : Boston Housing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/boston_housing.csv')
df
```

```python
# 1) 10번째 값 찾기
df.sort_values('crim', ascending = False)[:10] # 25.9406
```

```python
# 2) 값 변환
df.loc[df['crim'] >= 25.9406,'crim'] = 25.9406
```

```python
# 확인
df.sort_values('crim', ascending = False)[:10]
```

```python
# 2) 평균 구하기
df[df['age'] >= 80]['crim'].mean()
```

# 문제 1-2
- 데이터의 첫 번째 행부터 순서대로 80%까지의 데이터를 훈련 데이터로 추출 후 'total_bedrooms' 변수의 결측값을 'total_bedrooms' 변수의 중앙값으로 대체하고 대체 전의 'total_bedrooms'변수 표준편차 값의 차이의 절댓값을 구하시오.

```python
df = pd.read_csv('./data/california_housing.csv')
df
```

```python
train = df[:int(round(len(df)*0.8,0))]
train
```

```python
before = train['total_bedrooms'].std()
before
```

```python
train.loc[train['total_bedrooms'].isnull(),'total_bedrooms'] = df['total_bedrooms'].median()
```

```python
after = train['total_bedrooms'].std()
after
```

```python
abs(before - after)
```
